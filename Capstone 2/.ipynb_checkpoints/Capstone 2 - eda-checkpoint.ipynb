```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read cleaned data into pandas dataframe
filename = 'drug_use_data_cleaned.csv'
data = pd.read_csv( filename )
data.head()
```

```python
# Display countplot of ethnicities
# There appears to be a predominant amount of White individuals in the sample
# It is also unclear what "Other" consists of
_ = sns.catplot(x="Ethnicity", data=data, kind="count")
_.set_xticklabels(rotation=90)
plt.show()
```

```python
# Display countplot of "Country" to see if there is a correlation for significantly higher White participants
# Shows samples came from countries with predominantly White populations
# Also unclear what scope of countries "Other" covers
_ = sns.catplot(x="Country", data=data, kind="count")
_.set_xticklabels(rotation=90)
plt.show()
```

```python
# Display counts of different education levels
# Reorder values according to ascending education level for clearer visualization of distribution
# This data appears to be skewed left, which majority of the sample having a Professional certificate/diploma and higher
#Note: look at correlation between drug use and education
education_x = ["Left school before 16 years",
               "Left school at 16 years",
               "Left school at 17 years",
               "Left school at 18 years",
               "Professional certificate/diploma",
               "Some college or university, no certificate or degree",
               "University degree",
               "Masters degree",
               "Doctorate degree"]
_ = sns.catplot(x="Education", data=data, kind="count", order=education_x)
_.set_xticklabels(rotation=90)
plt.show()
```

```python
ed_ecs_data = data[["Education", "Ecstasy"]]
ed_ecs_grouped = ed_ecs_data.groupby("Education")["Ecstasy"].apply(lambda x: x.value_counts() )#normalize=True) )
ed_ecs_grouped.unstack().plot.bar(stacked=True);
```

```python
# Look at countplot of Semer (imaginary drug) (ordered in descending value by count)
# While majority of the participants truthfully answered (CL0 = Never used), 
# a small number of participants indicated use of the drug to some degree (CL1-CL4)
semer_col = data["Semer"]
_ = sns.catplot(x="Semer", data=data, kind="count", order=semer_col.value_counts().index)
plt.show()
```
